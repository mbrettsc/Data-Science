```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv('Training_knight.csv')
val_df = pd.read_csv('Validation_knight.csv')
test_df = pd.read_csv('../Test_knight.csv')
```

```python
X_train = train_df.drop(columns=['knight'])
y_train = train_df['knight']

X_val = val_df.drop(columns=['knight'])
y_val = val_df['knight']

X_test = test_df
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
dtc = DecisionTreeClassifier(random_state=42)
logreg = LogisticRegression(random_state=42, max_iter=1000)
```

```python
voting_clf = VotingClassifier(estimators=[
    ('KNN', knn), 
    ('DTC', dtc), 
    ('LOGREG', logreg)], voting='hard')
```

```python
voting_clf.fit(X_train_scaled, y_train)
```

```python
y_val_pred = voting_clf.predict(X_val_scaled)
f1 = f1_score(y_val, y_val_pred, pos_label='Jedi')
print(f"F1-Score on validation set: {f1:.4f}")
```

```python
y_test_pred = voting_clf.predict(X_test_scaled)

with open('Voting.txt', 'w') as f:
    for pred in y_test_pred:
        f.write(f"{pred}\n")
```
