```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../Train_knight.csv')
```

```python
df_numerical = df.drop(columns=['knight'])

variances = df_numerical.var()
total_variance = np.sum(variances)
variance_percentages = variances / total_variance * 100
sorted_variances = variance_percentages.sort_values(ascending=False)
cumulative_variances = sorted_variances.cumsum()
```

```python
print("Variances (Percentage):")
print(variance_percentages.values)

print("\nCumulative Variances (Percentage):")
print(cumulative_variances.values)
```

```python
plt.plot(np.arange(1, len(cumulative_variances) + 1), cumulative_variances)
plt.xlabel('Number of components')
plt.ylabel('Explained variance (%)')
plt.show()
```
