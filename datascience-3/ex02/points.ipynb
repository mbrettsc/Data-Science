```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/mbrettsc/code/data-science/datascience-3/Train_knight.csv')
```

```python
np.random.seed(42)
random_columns = np.random.choice(df.columns[:-1], 4, replace=False)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

sns.scatterplot(data=df, x=random_columns[0], y=random_columns[1], hue="knight", palette='bright', ax=axes[0, 0])
sns.scatterplot(data=df, x=random_columns[2], y=random_columns[1], hue="knight", palette='bright', ax=axes[0, 1])
sns.scatterplot(data=df, x=random_columns[2], y=random_columns[3], ax=axes[1, 0], color='green')
sns.scatterplot(data=df, x=random_columns[0], y=random_columns[3], ax=axes[1, 1], color='green')
plt.tight_layout()
plt.show()
```
