```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/mbrettsc/code/data-science/datascience-3/Test_knight.csv')
```

```python
fig, axes = plt.subplots(nrows=6, ncols=5, figsize=(15, 12))
axes = axes.flatten()

for i, col in enumerate(df.columns):
    axes[i].hist(df[col], bins=30, label='Knight', color='green', alpha=0.5)
    axes[i].legend()
    axes[i].set_title(col)

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('/Users/mbrettsc/code/data-science/datascience-3/Train_knight.csv')
```

```python
fig, axes = plt.subplots(nrows=6, ncols=5, figsize=(15, 12))
axes = axes.flatten()

for i, col in enumerate(df.columns[:-1]):
    axes[i].hist(df[df['knight'] == 'Sith'][col], bins=30, alpha=0.3, label='Sith', color='red')
    axes[i].hist(df[df['knight'] == 'Jedi'][col], bins=30, alpha=0.5, label='Jedi', color='blue')
    axes[i].set_title(col)
    axes[i].legend()

plt.tight_layout()
plt.show()
```
