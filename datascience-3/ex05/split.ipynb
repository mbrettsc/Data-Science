```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
def split_data(input_file, test_size=0.2, random_state=42):
    df = pd.read_csv(input_file)

    train_df, validation_df = train_test_split(df, test_size=test_size, random_state=random_state)

    train_df.to_csv('Training_knight.csv', index=False)
    validation_df.to_csv('Validation_knight.csv', index=False)

    return train_df, validation_df
```

```python
input_file = 'Train_knight.csv'

train_df, validation_df = split_data(input_file)

print(f"Training data shape: {train_df.shape}")
print(f"Validation data shape: {validation_df.shape}")

print("Training data preview:")
display(train_df.head())

print("Validation data preview:")
display(validation_df.head())
```
